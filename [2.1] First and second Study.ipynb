## Chi-square test USA

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (official results)
observed = np.array([57.94, 42.06])

# Expected frequencies (predicted results)
expected = np.array([57.39, 42.61])

# Perform the chi-square test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output the results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Determine significance
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The discrepancies are statistically significant.")
else:
    print("Fail to reject the null hypothesis: The discrepancies are not statistically significant.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare

# Observed frequencies (official results)
observed = np.array([57.94, 42.06])

# Expected frequencies (predicted results)
expected = np.array([57.39, 42.61])

# Perform the chi-square test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output the results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Determine significance
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The discrepancies are statistically significant.")
else:
    print("Fail to reject the null hypothesis: The discrepancies are not statistically significant.")

# Create bar charts for observed and expected frequencies
labels = ['Category 1', 'Category 2']
x = np.arange(len(labels))  # the label locations

# Set the width of the bars
width = 0.35  

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, observed, width, label='Observed', color='blue')
bars2 = ax.bar(x + width/2, expected, width, label='Expected', color='orange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frequencies')
ax.set_title('Observed vs Expected Frequencies')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Add value labels on top of the bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)

# Show the plot
plt.tight_layout()
plt.show()
```

# Predictive stats compare to Official results USA study

```python
# Predictive stats compare to Official results
# Differences and accuracy

# Official and predicted results
official_results = {
    "Donald J. Trump": 57.94,
    "Kamala Harris": 42.06
}

predicted_results = {
    "Donald J. Trump": 57.39,
    "Kamala Harris": 42.61
}

# Calculate differences
differences = {}
for candidate in official_results:
    differences[candidate] = abs(predicted_results[candidate] - official_results[candidate])

# Print results
for candidate in official_results:
    official = official_results[candidate]
    predicted = predicted_results[candidate]
    difference = differences[candidate]
    
    if predicted > official:
        sign = "+"
    else:
        sign = "-"
    
    print(f"{candidate}:")
    print(f"  Official Percentage: {official:.2f}%")
    print(f"  Predicted Percentage: {predicted:.2f}%")
    print(f"  Difference = {sign}{difference:.2f}%")
    print()
```

# Chi Square test Ecuador

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create the data frames
predictor_data = {
    'Candidate': ['Daniel Noboa', 'Luisa Gonzales', 'Andrea Gonzales', 'Pedro Granja', 
                  'Francesco Tabacchi', 'Leonidas Iza', 'Carlos Rabascall', 'Luis Tilleria', 
                  'Jimmy Jairala', 'Henry Cucalon', 'Ivan Saquicela', 'Jorge Escala', 
                  'Victor Arauz', 'Wilson Gomez', 'Henry Kronfle', 'Juan Cueva'],
    'Percentage': [54.65, 29.43, 3.50, 3.08, 2.96, 2.63, 1.37, 0.59, 0.43, 0.32, 
                   0.21, 0.20, 0.20, 0.16, 0.16, 0.11]
}

official_data = {
    'Candidate': ['DANIEL NOBOA AZIN', 'LUISA GONZALEZ', 'LEONIDAS IZA', 
                  'ANDREA GONZALEZ', 'HENRY KRONFLE KOZHAYA', 'PEDRO GRANJA', 
                  'JIMMY JAIRALA VALLAZZA', 'JORGE ESCALA', 'HENRY CUCALON', 
                  'LUIS FELIPE TILLERIA', 'FRANCESCO TABACCHI', 'VICTOR ARAUS', 
                  'CARLOS RABASCALL', 'WILSON GOMEZ', 'JUAN IVAN CUEVA', 
                  'IVAN SAQUICELA'],
    'Percentage': [44.17, 44.00, 5.25, 2.69, 0.71, 0.53, 0.40, 0.39, 0.36, 
                   0.32, 0.26, 0.22, 0.22, 0.18, 0.17, 0.12]
}

df_predictor = pd.DataFrame(predictor_data)
df_official = pd.DataFrame(official_data)

# Normalize the percentages to counts (assuming total votes = 100 for simplicity)
total_votes = 100
df_predictor['Count'] = (df_predictor['Percentage'] / 100) * total_votes
df_official['Count'] = (df_official['Percentage'] / 100) * total_votes

# Merge the data frames on Candidate
merged_df = pd.merge(df_predictor, df_official, how='outer', on='Candidate', suffixes=('_predictor', '_official'))

# Fill NaN values with 0 for counts
merged_df.fillna(0, inplace=True)

# Create a contingency table
contingency_table = merged_df[['Count_predictor', 'Count_official']].values

# Run the Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Output the results
print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two distributions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two distributions.")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Define observed counts based on the assumed sample size
sample_size = 1000
observed_counts = np.array([[989, 11],  # USA Accuracy Campaign Winner
                            [688, 312]])  # Overall Accuracy of Ecuador Model

# Step 2: Create a contingency table
contingency_table = pd.DataFrame(observed_counts, 
                                  columns=['Success', 'Failure'], 
                                  index=['USA Accuracy', 'Ecuador Model'])

# Step 3: Perform Chi-Square test
chi2_stat, p_value, dof, expected_counts = chi2_contingency(contingency_table)

# Step 4: Output results
print("Chi-Square Statistic:", chi2_stat)
print("P-Value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Counts:\n", expected_counts)

# Step 5: Determine significance
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The discrepancies are statistically significant.")
else:
    print("Fail to reject the null hypothesis: The discrepancies are not statistically significant.")
```
